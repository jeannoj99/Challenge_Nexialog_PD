```python
import dash
from dash import dcc, html
from dash.dependencies import Input, Output

# Initialiser l'application Dash
app = dash.Dash(__name__, suppress_callback_exceptions=True)

background_color = '#04153E'
text_color = '#FFFFFF'
menu_color = '#142345'
shadow_color = 'rgba(0, 0, 0, 0.3)'
selected_button_color = '#0F1B37' 

#les pages
new_page_layout = html.Div([
    html.Div(id='begin-page-content', children=[
        html.H1('Application', style={'textAlign': 'center', 'color': text_color}),
    ], style={'margin-left': '220px', 'padding': '20px', 'background-color': background_color, 'height': '100vh', 'width': 'calc(100vw - 220px)'}),
    html.Div([
        html.Ul([
            html.Li([
                html.H3('Navigation', style={'color': text_color}),
                html.Button([html.I(className="fas fa-home"), html.Span("Home", style={'margin-left': '10px'})], id='home-button', n_clicks=0, style={'display': 'block', 'margin': '0', 'padding': '10px 20px', 'width': '100%', 'color': text_color, 'background-color': menu_color, 'border': 'none', 'text-align': 'left'}),
                html.Button([html.I(className="fas fa-chart-line"), html.Span("Dashboard", style={'margin-left': '10px'})], id='dashboard-button', n_clicks=0, style={'display': 'block', 'margin': '0', 'padding': '10px 20px', 'width': '100%', 'color': text_color, 'background-color': menu_color, 'border': 'none', 'text-align': 'left'})
            ]),
            html.Li([
                html.H3('Données', style={'color': text_color}),
                html.Button([html.I(className="fas fa-database"), html.Span("Données", style={'margin-left': '10px'})], id='data-button', n_clicks=0, style={'display': 'block', 'margin': '0', 'padding': '10px 20px', 'width': '100%', 'color': text_color, 'background-color': menu_color, 'border': 'none', 'text-align': 'left'})
            ])
        ], style={'list-style-type': 'none', 'padding': '0'})
    ], style={'position': 'fixed', 'left': '0', 'top': '0', 'bottom': '0', 'width': '200px', 'background-color': menu_color, 'padding': '20px', 'box-shadow': f'5px 5px 15px {shadow_color}'})
], style={'overflow': 'hidden'})  # Le débordement est caché pour empêcher le défilement de la page

# Layout initial
app.layout = new_page_layout

# Callback pour afficher le contenu de la nouvelle page ou revenir à la page d'accueil
@app.callback(
    Output('begin-page-content', 'children'),
    [Input('home-button', 'n_clicks'),
     Input('dashboard-button', 'n_clicks'),
     Input('data-button', 'n_clicks')]
)
def update_page_content(home_clicks, dashboard_clicks, data_clicks):
    ctx = dash.callback_context
    if not ctx.triggered:
        return html.H1('Application', style={'textAlign': 'center', 'color': text_color})  # La page affiché des le debut avec le mot "Application"

    button_id = ctx.triggered[0]['prop_id'].split('.')[0]

    if button_id == 'home-button' and home_clicks > 0:
        return html.Div([
            html.H1('Welcome', style={'textAlign': 'center', 'color': text_color})
        ], style={'background-color': background_color, 'height': '100vh', 'width': 'calc(100vw - 220px)'})
    elif button_id == 'dashboard-button' and dashboard_clicks > 0:
        # Code pour afficher la page Dashboard
        return html.Div([
            html.H1('Dashboard', style={'textAlign': 'center', 'color': text_color})
            # Ajoutez ici le contenu du tableau de bord
        ], style={'background-color': background_color, 'height': '100vh', 'width': 'calc(100vw - 220px)'})
    elif button_id == 'data-button' and data_clicks > 0:
        # Code pour afficher la page Données
        return html.Div([
            html.H1('Données', style={'textAlign': 'center', 'color': text_color})
            # Ajoutez ici le contenu relatif aux données
        ], style={'background-color': background_color, 'height': '100vh', 'width': 'calc(100vw - 220px)'})
    else:
        return html.H1('Application', style={'textAlign': 'center', 'color': text_color})  # Afficher le mot "Application" avant que l'utilisateur appuie sur un bouton

# Callback pour mettre à jour le style des boutons en fonction de leur état
@app.callback(
    [Output('home-button', 'style'),
     Output('dashboard-button', 'style'),
     Output('data-button', 'style')],
    [Input('home-button', 'n_clicks'),
     Input('dashboard-button', 'n_clicks'),
     Input('data-button', 'n_clicks')]
)
def update_button_styles(home_clicks, dashboard_clicks, data_clicks):
    #création des boutons
    home_style = {'display': 'block', 'margin': '0', 'padding': '10px 20px', 'width': '100%', 'color': text_color, 'background-color': menu_color, 'border': 'none', 'text-align': 'left'}
    dashboard_style = {'display': 'block', 'margin': '0', 'padding': '10px 20px', 'width': '100%', 'color': text_color, 'background-color': menu_color, 'border': 'none', 'text-align': 'left'}
    data_style = {'display': 'block', 'margin': '0', 'padding': '10px 20px', 'width': '100%', 'color': text_color, 'background-color': menu_color, 'border': 'none', 'text-align': 'left'}

    #foncer le bouton selectionné
    ctx = dash.callback_context
    if ctx.triggered:
        button_id = ctx.triggered[0]['prop_id'].split('.')[0]
        if button_id == 'home-button':
            home_style['background-color'] = selected_button_color
        elif button_id == 'dashboard-button':
            dashboard_style['background-color'] = selected_button_color
        elif button_id == 'data-button':
            data_style['background-color'] = selected_button_color

    return home_style, dashboard_style, data_style

if __name__ == '__main__':
    app.run_server(debug=True, port=8088)
```
